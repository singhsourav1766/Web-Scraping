Q1: Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You

jobs data.

2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúSkill, Designations, Companies‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the 
location‚Äù field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results youget.

```python
# Let's Install automated chrome browser and connect it to web driver
driver= webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys('Data Analyst')    #Enter Data Analyst in Designation

location=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys('Bangalore')    #Enter location as Bangalore
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title =[]
job_location =[]
company_name =[]
experience_required =[]
```

```python
title_tags=driver.find_elements(By.XPATH, '//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    job_title.append(i.text)

location_tags=driver.find_elements(By.XPATH, '//span[@class="ellipsis fleft locWdth"]')
for i in location_tags[0:10]:
    job_location.append(i.text)

company_tags=driver.find_elements(By.XPATH, '//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    company_name.append(i.text)

experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experience_tags[0:10]:
    experience_required.append(i.text)

# Printing the length of each list
print(len(job_title))
print(len(job_location))
print(len(company_name))
print(len(experience_required))
```

```python
dfda=pd.DataFrame({'Title':job_title,'Location':job_location,'Company_names':company_name,'Experiance':experience_required})
dfda.index=range(1,len(dfda)+1)
dfda
```

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
# Let's Install automated chrome browser and connect it to web driver
driver=webdriver.Chrome()
```

```python
driver.get('https://www.azquotes.com/')
```

```python
quotes=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]')
quotes.click()
```

```python
Quote=[]
Author=[]
Type_of_quotes=[]
```

```python
start=0
end=10
for page in range(start, end):
    quote_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
    
    for i in quote_tag:
        Quote.append(i.text)
    author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]')
    
    for i in author_tag:
        Author.append(i.text)
    type_of_quotes_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    
    for i in type_of_quotes_tag:
        Type_of_quotes.append(i.text)
    
    nexts=driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]')
    nexts.click()
    time.sleep(3)
    
```

```python
print(len(Quote))
print(len(Author))
print(len(Type_of_quotes))
```

```python
dfq=pd.DataFrame({'Quotes':Quote,'Author':Author,'Type Of Quotes':Type_of_quotes})
dfq.index=range(1,len(dfq)+1)
dfq
```
